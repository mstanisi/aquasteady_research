```python
import requests
import os
os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd
import shapely
```

```python
BingMapsKey = "Ao9j3s8EKOs7DfiBZ9TkaSUqt6qJS089cn9CO0ObFMsxPC4Wn6aXSbpXNJp1EiIG"
lat = 40.73832129328166
lon = -73.99897111941965
max_time = 30
time_unit = 'minute'
date_time = '03/01/2025T05:42:00'
```

```python
url = f"https://dev.virtualearth.net/REST/v1/Routes/Isochrones?waypoint={lat},{lon}&maxtime={max_time}&timeUnit={time_unit}&dateTime={date_time}&key={BingMapsKey}"
```

```python
url
```

```python
data = requests.get(url).json()
```

```python
coords = data['resourceSets'][0]['resources'][0]['polygons'][0]['coordinates'][0]
```

```python
points = [[y, x] for x, y in coords]
```

```python
polygon = shapely.geometry.Polygon(points)
```

```python
isochrome = gpd.GeoDataFrame(geometry=[polygon], crs=4326)
```

```python
isochrome.explore()
```
